# GLOBAL MAP - LEVEL 0 DEMO

```python
# --- GLOBAL MAP --- INCLUDES ROOM 0 - Pre-Course

# 0 = Locked/Empty, 1 = Unlocked/Available

world_map = {
    "R0": {
        "name": "GIT INIT: SNAKEPIT",
        "tagline": "LOG: Snakes crawling everywhere! You need to get out of here... You're late for your IRONHACK BOOTCAMP that starts today!",
        "objects": {
            "anaconda": {
                "type": "DECOY", 
                "msg": "LOG: 'git status' -> FATAL: Anaconda is untracked and highly venomous. Do not touch!"
            },
            "book": {
                "type": "MAN", 
                "info": "LOG: 'Bumper Book of GIT Magick Spells'. The page on 'Invisible Ink' is ripped out."
            },
            "terrarium": {
                "type": "CACHE", 
                "item": "wood", 
                "status": 1,
                "msg_success": "LOG: 'git fetch origin terrarium' -> You found broken [wood] inside!",
                "msg_empty": "LOG: The terrarium is empty. The snakes have breached their repository."
            },
            "cauldron": {
                "type": "LOCK", 
                "req": "page", 
                "reward": "potion",
                "status": 0, 
                "locked_msg": "LOG: The cauldron bubbles ominously. It awaits a 'git commit' of the right ingredient.",
                "unlock_msg": "LOG: 'git commit page' -> Success! The ink clears, revealing the repair spell. You receive: [potion]."
            },
            "ladder": {
                "type": "DOOR", 
                "req": ["wood", "potion"], 
                "target": "R1", 
                "status": 0,
                "locked_msg": "LOG: 'git push origin ladder' -> REJECTED. Missing dependencies. You need broken [wood] and a repair [potion]."
            }
        }
    },
    "R1": {
        "name": "LEVEL 1: IRONHACK BOOTCAMP",
        "tagline": "LOG: You slithered out of the pit and made it to class!",
        "objects": {} 
        # R1-R3 logic goes here later.
    }
}

# --- PLAYER STATE ---
state = {
    "current_room": "R0",
    "inventory": ["page"], # Player starts with KeyA in fist
    "game_over": 0
}

# --- GLOBAL ENGINE LOGIC ---
def show_room():
    room = world_map[state["current_room"]]
    print(f"\n{'='*40}\n{room['name']}\n{'='*40}")
    print(room["tagline"])

def interact(target):
    room = world_map[state["current_room"]]
    
    if target not in room["objects"]:
        print(f"LOG: Error 404 - '{target}' not found in current directory.")
        return

    obj = room["objects"][target]
    
    # 1. DECOY & MAN
    if obj["type"] in ["DECOY", "MAN"]:
        msg_key = "msg" if obj["type"] == "DECOY" else "info"
        print(obj[msg_key])

    # 2. CACHE
    elif obj["type"] == "CACHE":
        if obj["status"] == 1:
            state["inventory"].append(obj["item"])
            obj["status"] = 0 
            print(obj["msg_success"])
        else:
            print(obj["msg_empty"])

    # 3. LOCK (Cauldron)
    elif obj["type"] == "LOCK":
        if obj["status"] == 1:
            print("LOG: The cauldron is empty. You already compiled the potion.")
        elif obj["req"] in state["inventory"]:
            state["inventory"].append(obj["reward"])
            obj["status"] = 1
            print(obj["unlock_msg"])
        else:
            print(obj["locked_msg"])

    # 4. DOOR (Ladder)
    elif obj["type"] == "DOOR":
        # Check if ALL requirements are in inventory
        if all(req in state["inventory"] for req in obj["req"]):
            state["current_room"] = obj["target"]
            print(f"LOG: 'git merge wood potion' -> Ladder repaired! Slithering up to {obj['target']}...")
        else:
            print(obj["locked_msg"])

# --- MAIN LOOP ---
def play():
    # Only show full room intro once per room entry
    current_tracker = ""
    
    while state["game_over"] == 0:
        if current_tracker != state["current_room"]:
            show_room()
            current_tracker = state["current_room"]
            
        # Stop the MVP loop if we successfully hit R1
        if state["current_room"] == "R1":
            print("\nLOG: --- END OF ROOM 0 DEMO ---")
            break

        # Dynamic prompt showing visible items
        visible_items = list(world_map[state["current_room"]]["objects"].keys())
        cmd = input(f"\nWhat next? (Visible: {visible_items})\n> ").strip().lower()
        
        if cmd == "quit": 
            print("LOG: Game aborted.")
            break
            
        interact(cmd)

# EXECUTE
if __name__ == "__main__":
    play()
```
